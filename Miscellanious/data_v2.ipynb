```python
import pandas as pd
import os
```

```python
nox_data = pd.read_csv('datasets/no2.csv')
suicide_data = pd.read_csv('datasets/suicide.csv')
sunshine_data = pd.read_csv('datasets/sunshine_hours.csv')
```

```python
nox_data.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(nox_data):
    # Drop column: 'City'
    nox_data = nox_data.drop(columns=['City', 'Region'])
    nox_data = nox_data.rename(columns={'Country Sort descending': 'country'})
    return nox_data

nox_data_clean = clean_data(nox_data.copy())
nox_data_clean.head()
```

```python
suicides_clean = suicide_data.groupby(['year', 'country'])['suicides_no'].sum().reset_index()

suicides_clean.sort_values(['year', 'country'])
```

```python
#sunshine_data_clean = sunshine_data.loc[:, ['Country', 'Year']].groupby('Country').mean()

#sunshine_data_clean.rename({'Year': 'Annual mean'}, inplace=True)
#sunshine_data_clean = sunshine_data_clean.reset_index().rename({'Year': 'Annual mean', 'Country': 'country'})
```

```python
#sunshine_data_clean = sunshine_data.loc[:, ['Country', 'Year']].groupby('Country').mean()
#sunshine_data_clean = sunshine_data_clean.reset_index().rename({'Year': 'Annual mean', 'Country': 'country'})
sunshine_data_clean = sunshine_data.loc[:, ['Country', 'Year']].groupby('Country').mean().reset_index()
sunshine_data_clean = sunshine_data_clean.rename(columns = {'Country': 'country', 'Year': 'annual mean'})
sunshine_data_clean
```

```python
datasets_2 = {
    'sunshine_data_clean': sunshine_data_clean,
    'nox_data_clean': nox_data_clean,
    'suicides_clean': suicides_clean
}

for name, dataset in datasets_2.items():
    dataset.to_csv(f'processed_datasets/{name}.csv', index=False)
```
