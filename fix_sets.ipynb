```python
import pandas as pd
```

```python
elec = pd.read_csv('datasets/electricity.csv')
pop = pd.read_csv('datasets/population.csv')
pt = pd.read_csv('datasets/public_transport.csv')
```

```python
elec = elec.loc[:, ['ISO3', 'NAME_0', 'SDG711pct']]
elec.rename(columns = {'NAME_0': 'Country', 'SDG711pct': 'Electricity access of pop (%)'}, inplace = True)
```

```python
pop = pop.loc[:, ['Country Name', 'Country Code'] + [str(year) for year in range(1990, 2024)]]
```

```python
pt.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(pt):
    # Drop columns: 'OID_  ', 'UC_NM_MN                              ' and 3 other columns
    pt = pt.drop(columns=['OID_  ', 'UC_NM_MN                              ', 'UC_NM_LST                                                                                                                                                                                                                                                                                                                                                                                                                                             ', 'Shape_Length       ', 'Shape_Area'])
    return pt

pt_clean = clean_data(pt.copy())
pt_clean.head()
```

```python
pt_clean.columns = pt_clean.columns.str.strip()
```

```python
pt_clean = pt_clean[pt_clean['Access_POP'] != -9999]
#pt_clean = pt_clean.drop(columns=['SDG1121pct'])

summary = pt_clean.groupby(['CTR_MN_ISO','CTR_MN_NM'] ).agg({'Total_POP': 'sum', 'Access_POP': 'sum'}).reset_index()
summary.rename(columns={'CTR_MN_ISO': 'ISO3', 'CTR_MN_NM': 'country'}, inplace=True)
summary['pct'] = (summary.Access_POP / summary.Total_POP) *100
```

```python
summary.drop(columns = ['Total_POP', 'Access_POP'], inplace=True)
```
