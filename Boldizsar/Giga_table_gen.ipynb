```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
gini_data_df = pd.read_csv('./processed_datasets/gini_data.csv')
gini_data_df
```

```python
# Initialize an empty list to store DataFrames
df_list = []

for year in range(1990, 2021):  # Assuming the range of years is from 1991 to 2023
    year_column = str(year)
    
    # Check if the year exists as a column in the dataframe
    if year_column in gini_data_df.columns:
        # Create a new DataFrame with 'country', 'alpha3', and the value of that year
        df_year = gini_data_df[['country', 'alpha3', year_column]].copy()
        
        # Rename the column to 'gini' instead of 'value'
        df_year = df_year.rename(columns={year_column: 'gini'})
        
        # Add a new column 'year' to store the current year
        df_year['year'] = year
        
        # Rearrange the columns to ['country', 'year', 'alpha3', 'gini']
        df_year = df_year[['country', 'year', 'alpha3', 'gini']]
        
        # Append the DataFrame to the list
        df_list.append(df_year)

# Concatenate all the DataFrames in the list into a single DataFrame
merged_gini_data = pd.concat(df_list, ignore_index=True)

# Optionally, inspect the merged DataFrame
print(merged_gini_data.head())
```

```python
population_pd = pd.read_csv('./processed_datasets/population.csv')
population_pd
```

```python
# Initialize an empty list to store DataFrames
df_list = []

for year in range(1990, 2021):  # Assuming the range of years is from 1990 to 2020
    year_column = str(year)
    
    # Check if the year exists as a column in the dataframe
    if year_column in population_pd.columns:
        # Create a new DataFrame with 'Country Name', 'Country Code', and the value of that year
        df_year = population_pd[['Country Name', 'Country Code', year_column]].copy()
        
        # Rename the columns as per the required structure
        df_year = df_year.rename(columns={
            'Country Name': 'country',
            'Country Code': 'alpha3',
            year_column: 'population'  # Rename to 'population' instead of 'value'
        })
        
        # Add a new column 'year' to store the current year
        df_year['year'] = year
        
        # Rearrange the columns to ['country', 'year', 'alpha3', 'population']
        df_year = df_year[['country', 'year', 'alpha3', 'population']]
        
        # Append the DataFrame to the list
        df_list.append(df_year)

# Concatenate all the DataFrames in the list into a single DataFrame
merged_population_data = pd.concat(df_list, ignore_index=True)

# Optionally, inspect the merged DataFrame
print(merged_population_data.head())
```

```python
poverty_data_pd = pd.read_csv('./processed_datasets/poverty_data.csv')
poverty_data_pd
```

```python
# Initialize an empty list to store DataFrames
df_list = []

for year in range(1990, 2021):  # Assuming the range of years is from 1990 to 2020
    year_column = str(year)
    
    # Check if the year exists as a column in the dataframe
    if year_column in poverty_data_pd.columns:
        # Create a new DataFrame with 'Country Name', 'Country Code', and the value of that year
        df_year = poverty_data_pd[['Country Name', 'Country Code', year_column]].copy()
        
        # Rename the columns as per the required structure
        df_year = df_year.rename(columns={
            'Country Name': 'country',
            'Country Code': 'alpha3',
            year_column: 'poverty'  
        })
        
        # Add a new column 'year' to store the current year
        df_year['year'] = year
        
        # Rearrange the columns to ['country', 'year', 'alpha3', 'population']
        df_year = df_year[['country', 'year', 'alpha3', 'poverty']]
        
        # Append the DataFrame to the list
        df_list.append(df_year)

# Concatenate all the DataFrames in the list into a single DataFrame
merged_poverty_data = pd.concat(df_list, ignore_index=True)

# Optionally, inspect the merged DataFrame
print(merged_poverty_data.head())
```

```python
suicides_df = pd.read_csv('./processed_datasets/suicides_clean.csv')
print(suicides_df)
```

```python
# Filter the DataFrame for the years between 1990 and 2020 (inclusive)
filtered_suicides_df = suicides_df[(suicides_df['year'] >= 1990) & (suicides_df['year'] <= 2020)]

# Optionally, inspect the filtered DataFrame
print(filtered_suicides_df.head())

```

```python
# Start with the first DataFrame (e.g., gini data)
merged_df = merged_gini_data

# Now merge it iteratively with other DataFrames, specifying suffixes to avoid duplicates
merged_df = pd.merge(merged_df, merged_literacy_data, on=['country', 'year'], how='outer', suffixes=('_gini', '_literacy'))
merged_df = pd.merge(merged_df, merged_population_data, on=['country', 'year'], how='outer', suffixes=('', '_population'))
merged_df = pd.merge(merged_df, merged_poverty_data, on=['country', 'year'], how='outer', suffixes=('', '_poverty'))
merged_df = pd.merge(merged_df, filtered_suicides_df, on=['country', 'year'], how='outer', suffixes=('', '_suicides'))

# Optionally inspect the merged DataFrame
merged_df.dropna().size
```
