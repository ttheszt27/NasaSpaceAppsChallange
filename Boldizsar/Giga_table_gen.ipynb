```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
gini_data_df = pd.read_csv('./processed_datasets/gini_data.csv')
gini_data_df
```

```python
for year in range(1990, 2021):  # Assuming the range of years is from 1991 to 2023
    year_column = str(year)
    
    # Check if the year exists as a column in the dataframe
    if year_column in gini_data_df.columns:
        # Create a new DataFrame with 'country', 'alpha3', and the value of that year
        df_year = gini_data_df[['country', 'alpha3', year_column]].copy()
        
        # Rename the column to 'value' for that year
        df_year = df_year.rename(columns={year_column: 'value'})
        
        # Add a new column 'year' to store the current year
        df_year['year'] = year
        
        # Rearrange the columns to ['country', 'year', 'alpha3', 'value']
        df_year = df_year[['country', 'year', 'alpha3', 'value']]
        
        # Dynamically create a new DataFrame variable
        globals()[f'gini_data_df_{year}'] = df_year

        # Optionally, print or inspect the result for a specific year
        print(f'gini_data_df_{year}')
        print(df_year.head())
```

```python
suicides_df = pd.read_csv('./processed_datasets/suicides_clean.csv')
print(suicides_df)
```

```python

# Get unique years in the dataset
unique_years = suicides_df['year'].unique()

# Loop through each unique year and create a DataFrame
for year in unique_years:
    # Filter the DataFrame for the specific year
    df_year = suicides_df[suicides_df['year'] == year].copy()
    
    # Dynamically create a new DataFrame variable for each year
    globals()[f'suicides_df_{year}'] = df_year
    
    # Optionally, print or inspect the result for a specific year
    print(f'\nsuicides_df_{year}')
    print(df_year.head())
```
