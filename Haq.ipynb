```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('./kaggle/input/healthcare-access-and-quality-index.csv')

# Rename columns for easier access
df.columns = ['Entity', 'Code', 'Year', 'HAQ_Index']

# Define the countries of interest
countries_of_interest = ["Hungary", "Slovakia", "Romania", "Poland", "Croatia", "Serbia", "Montenegro", "Albania", 
                         "Czechia", "Bulgaria", "Greece", "North Macedonia", "Moldova", "Lithuania", "Latvia", 
                         "Germany", "Austria", "Switzerland", "Slovenia", "Sweden", "Estonia"]

# Filter the DataFrame to include only these countries
df_filtered = df[df['Entity'].isin(countries_of_interest)]

# Set up the plot
plt.figure(figsize=(14, 8))

# Plot each country
for country in countries_of_interest:
    # Subset the data for the specific country
    country_data = df_filtered[df_filtered['Entity'] == country]
    
    # Plot data
    plt.plot(country_data['Year'], country_data['HAQ_Index'], label=country)

# Add labels and title
plt.title('HAQ Index for Selected Countries Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('HAQ Index (IHME 2017)', fontsize=12)

# Add legend outside the plot
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.tight_layout()
plt.show()
```
